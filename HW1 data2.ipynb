```python
train = pd.read_csv("data2/train.csv")
test = pd.read_csv("data2/test.csv")
sample = pd.read_csv("data2/sample_submission.csv")
```

```python
train.head(5)
```

```python
train.drop("ID",inplace=True,axis=1)
id = test["ID"]
test.drop("ID",inplace=True,axis=1)
names = sample.columns.tolist()
sample.drop("ID",inplace=True,axis=1)
```

```python
y_col = sample.columns.tolist()

y_train = train[y_col]
X_train = train.drop(y_col,axis=1)

print(X_train.shape)
print(y_train.shape)
```

```python
X_train.head(10)
```

```python
x = torch.Tensor(X_train.values)
y = torch.Tensor(y_train.values)
```

```python
sum(y_train.iloc[:,0])/y_train.shape[0]
```

```python
device = torch.device('cpu')
N, D_in,  D_out = 76020, 369, 1
H = [500,1000,500,100]

model = torch.nn.Sequential(
    torch.nn.Linear(D_in, H[1]),
    torch.nn.LeakyReLU(),
    torch.nn.Linear(H[1], D_out)
    
)

l_r = 1e-9

for i in range(100):
    y_pred = torch.sigmoid(model(x))

    loss = torch.nn.functional.binary_cross_entropy(y_pred,y)
    print(loss)
    loss.backward()

    with torch.no_grad():
        for param in model.parameters():
            param -= l_r*param.grad
    model.zero_grad()
```

```python
X_test_2l = torch.Tensor(test.values)
y_pred_2l = model(X_test_2l)
#torch.nn.functional.softmax(y_pred_2l)
y_pred_2l = pd.DataFrame(y_pred_2l.detach()).astype("float")
y_pred_2l
```
